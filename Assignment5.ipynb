# Question no.1

#  Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed 
# and average_of_vehicle

```python
class Vehicle:
    def __init__(self,name,speed,avg):
        self.name_of_vehicle=name
        self.max_speed=speed
        self.average_of_vehicle=avg
        
```

```python
v1=Vehicle('Hero Honda',200,50)
```

```python
v1.average_of_vehicle
```

```python
v1.max_speed
```

```python
v1.name_of_vehicle
```

# Question no2

#  Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class. 
# Create a method named seating_capacity which takes capacity as an argument and returns the name of 
# the vehicle and its seating capacity

```python
class Car(Vehicle):
    def seating_capacity(self,capacity):
        self.seating_capa=capacity
        return self.seating_capa,self.name_of_vehicle
        
        
```

```python
c=Car('toyota',90,80)
```

```python
c.seating_capacity(90)
```

# Question no3

 # What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

```python
# Multiple inheritance means a class inherits properties and methods from more than one parent class.
```

```python
class Father :
    def skill1(self):
        print('Singing')
class Mother:
    def skill2(self):
        print('Cooking')
class Child(Father,Mother):
    def skill3(self):
        print('Dancing')

    
```

```python
c=Child()
```

```python
c.skill1()
```

```python
c.skill2()
```

```python
c.skill3()
```

# What is method overriding in python? Write a python code to demonstrate method overriding

```python
class Human:
    def eat(self):
        print('Human eat meal')

class Male(Human):
    def eat(self):
        print('Male eat meal')
class Female(Human):
    def eat(self):
        print('Female eat meal')
        
        
```

```python
h=Human()
m=Male()
f=Female()
```

```python
h.eat()
```
