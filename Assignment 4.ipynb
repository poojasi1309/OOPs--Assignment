# Explain class and object with respect to object oriented programming .give a suitable example.

````python

1. Class-
A class is a blueprint or template used to create objects. It defines:
Attributes (variables) → properties of the object
Methods (functions) → behavior of the object
Example:
A class Car can have:
Attributes: color, brand, speed
Methods: start(), stop(), accelerate()

2. Object
An object is an instance of a class. It represents a real-world entity and contains:
Actual values of attributes
Ability to use methods defined in the class.

Example:
If Car is a class, then:
car1 = BMW, red, 120 km/h
car2 = Audi, black, 150 km/h
These are objects.
````

#  Explain why the __init__() is uesed .Give a suitable example.

````python
The __init__() function is a constructor in Python. It is automatically called when an object of a class is created.


        
````

````python
# Example
class Student:
    def __init__(self,enroll,name,email):
        self.__enroll=enroll
        self.__name=name
        self.__email=email

    def show(self):
        print('Enroll',self.__enroll)
        print('Name',self.__name)
        print('email',self.__email)

s=Student(12,'ram','ram@12gmail.com')
s.show()
````

#  why self is used in OOps?

````python
self is used to refer to the current object (instance) of a class.
Access instance variables

It allows us to access variables that belong to a particular object.

class Student:
    def __init__(self, name):
        self.name = name   # instance variable

    def show(self):
        print(self.name)

2. Differentiate between local and instance variables
Without self, Python cannot differentiate between local variables and object variables.

class Demo:
    def __init__(self, value):
        self.value = value

3.Call other methods of the same class

Using self, one method can call another method inside the same class.

class Test:
    def method1(self):
        print("Hello")

    def method2(self):
        self.method1()   

4. Represents current object
Each object has its own data, and self helps refer to that specific object.
````

#  Question 5

# What is inheritance ? Give an example of each type of inheritance.

````python
Inheritance is a concept in Object-Oriented Programming where one class (child class) acquires the properties and methods of another class (parent class).
It helps in:

1. Code reusability
2. Reducing redundancy
3.Creating a hierarchical relationship

Types of Inheritance-

1. Single Inheritance

One child class inherits from one parent class.

Example-
class Parent:
    def fun1(self):
        print("This is Parent class")

class Child(Parent):
    def fun2(self):
        print("This is Child class")

obj = Child()
obj.fun1()
obj.fun2()

2. Multiple Inheritance-

One child class inherits from multiple parent classes.

```python
class Father:
    def skills1(self):
        print("Gardening")

class Mother:
    def skills2(self):
        print("Cooking")

class Child(Father, Mother):
    pass

obj = Child()
obj.skills1()
obj.skills2()

3. Multilevel Inheritance

A class inherits from another class, which is already inherited from another class.

```python
class Grandparent:
    def house(self):
        print("Grandparent's house")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

obj = Child()
obj.house()

4. Hierarchical Inheritance-

Multiple child classes inherit from the same parent class.

Example-
class Parent:
    def property(self):
        print("Parent property")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj2 = Child2()

obj1.property()
obj2.property()
````
